# Sorting and Rating (reytinglash )
- `.sort_index(axis=)` >> index buyicha sorting qilish
- `.sort_values(by='value')` >> qiymatlari buyicha soritng
-` ascending=False` >> Teskari tartibda sorting

`.rank()` metodi yordamida Df qatorlarini reytingini aniqlash mumkin. Reyting aniqlashda aynan qaysi ko'rsatkich (ustun) bo'yicvha reytingni aniqlash ko'rsatiladi


```python
df = pd.DataFrame(np.arange(1,16,2).reshape(2,4) , columns=['b', 'c', 'a','d'] , index=[2,1])
df
```

```python
df.sort_index(axis=1 , inplace=True)
```

```python
df.sort_index()
```

```python
# KEling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.sort_values(by='price' , ascending=False , inplace=True) # narxlar buyicha sortitng
```

```python
df.sort_index(inplace=True) # yuqorida narx buyicha sorting bulganda inplase  bn real df uzgargan edi keyin indexni qayta sorting qilindi
df
```

```python
df1 = df.iloc[: , [0,3,5]]
df1.head()
df1.index = df.brand
df1.sort_values('year' , ascending=False  , inplace=True)
df1
```

Reytinglash

`.rank()` metodi yordamida Df qatorlarini reytingini aniqlash mumkin. Reyting aniqlashda aynan qaysi ko'rsatkich (ustun) bo'yicvha reytingni aniqlash ko'rsatiladi

```python
# Moshinalarni mileage (yurgan kilometri) bo'yicha reyitnglaymiz
df['rank'] = df['mileage'].rank()
df
```

```python
# .rank() method parametri yordamida reyting o'lchash usulini tanlash mumkin (min, max, average, first, dense)
df['rank'] = df['year'].rank(method='average')
df.sort_values('year')
```

```python
df['rank'] = df['price'].rank(method='min')
df.sort_values('price')
```

```python
df['rank'] = df['color'].rank(method='average')
df.sort_values('color')
```
